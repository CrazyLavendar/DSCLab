```python
##########################################   SAMPLE DATA SET1   #############################################
df=pd.read_csv('dslab3_dataset1.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
x= df.name
y= df.cook_time
plt.figure(figsize=(25,10))
plt.plot(x,y)
plt.xlabel("Food Name")
plt.ylabel("Cooking Time")
plt.title('Food Making time')
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.bar(x,y)

plt.xlabel("Food Name")
plt.ylabel("Cooking Time")
plt.title('Food Making time')
plt.show()
```

```python
plt.figure(figsize=(25,10))
Preparation_time= df.prep_time
plt.hist(Preparation_time,30)
plt.xlabel("Preparation time of food")
plt.ylabel("No of dishes")
plt.title('Food Making time')
plt.show()
```

```python
plt.figure(figsize=(25,10))
region = df.region
Cook_time= df.cook_time
plt.pie(df.prep_time,labels = df.name )
plt.xlabel("Cooking time of food")
plt.ylabel("No of dishes")
plt.title('Food Making time')
plt.legend(loc =0)
plt.show()
```

```python
plt.figure(figsize=(25,10))
region = df.region
Cook_time= df.cook_time
plt.scatter(df.prep_time,df.cook_time )
plt.xlabel("Cooking time of food")
plt.ylabel("No of dishes")
plt.title('Food Making time')
plt.legend(loc =0)
plt.show()
```
